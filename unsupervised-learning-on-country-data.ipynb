```python

X = df.drop('country', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
wcss = []
silhouette_coefficients = []
for i in range(2, 16):
    model = KMeans(n_clusters=i)
    model.fit(X)
    wcss.append(model.inertia_) 
    silhouette_coefficients.append(silhouette_score(X, model.labels_))
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(121)
sns.lineplot(x=range(2, 16),y= wcss)
plt.title('wcss')
plt.subplot(122)
sns.lineplot(x=range(2, 16),y=silhouette_coefficients)
plt.title('silhouette_score')
```

```python
model = KMeans(n_clusters=4)
model.fit(X)
y = model.labels_
```

```python
silhouette_score(X, y)
```

```python
plt.figure(figsize=(20, 10))
from scipy.cluster import hierarchy
fig = hierarchy.dendrogram(hierarchy.linkage(y=X, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
silhouette_coefficients2 = []
for i in range(2, 16):
    hi = AgglomerativeClustering(n_clusters = i)
    hi.fit(X)
    silhouette_coefficients2.append(silhouette_score(X, hi.labels_))
```

```python
sns.lineplot(x=range(2, 16),y=silhouette_coefficients2)
```

```python
hi = AgglomerativeClustering(n_clusters = 2)
hi.fit(X)

y = hi.labels_
```
